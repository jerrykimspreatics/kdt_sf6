#### 공덕역 마커 표시하기

```python
import folium

# Map 클래스 사용, location-[경도, 위도], zoom_start-확대/축소
# 지도 스타일 - tiles="스타일"
# 지도 객체 생성
seoul_map = folium.Map(
    location=[37.54322, 126.951576], 
    zoom_start=15
)

# Marker 클래스 - icon : Icon 클래스 icon 속성-'star', 'arrow-up'
# Marker 매서드 - add_to(지도 객체)
folium.Marker(
    location=[37.54322, 126.951576], 
    icon = folium.Icon(icon='arrow-down', color='green'),
    zoom_start=15
).add_to(seoul_map)

seoul_map
```

```python
import folium

# Map 클래스 사용, location-[경도, 위도], zoom_start-확대/축소
# 지도 스타일 - tiles="스타일"
# 지도 객체 생성
seoul_map = folium.Map(
    location=[37.54322, 126.951576], 
    zoom_start=15
)

# Marker 클래스 - icon : Icon 클래스 icon 속성-'star', 'arrow-up'
# Marker 매서드 - add_to(지도 객체)
folium.Marker(
    location=[37.54322, 126.951576], 
    icon = folium.Icon(icon='arrow-down', color='green'),
    popup = "공덕역",
    zoom_start=15
).add_to(seoul_map)

folium.CircleMarker(
    location=[37.54322, 126.951576], 
    radius = 100,  #반지름 100px
    color = "#ff0",  #16진수 - 노란색
).add_to(seoul_map)

seoul_map
```
