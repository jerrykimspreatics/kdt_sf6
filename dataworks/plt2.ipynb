###### 자료의 분포 상태를 직사각형 모양의 막대그래프로 나타낸 것으로 데이터의 빈도에 따라 높이가 결정됨
###### hist(데이터리스트)

```python
import matplotlib.pyplot as plt
import random

# plt.hist([1, 1, 2, 3, 4, 5, 5, 5, 5, 7])

# 주사위 던지기
dice = []
for i in range(100):
    dice.append(random.randint(1, 6))
print(dice)

plt.hist(dice, bins=6) # 가로축의 구간의 개수
plt.show()
```

```python
 # 최고 기온만 리스트에 담기
import csv
import matplotlib.pyplot as plt

# with open("datas/temp_seoul.csv", "r", encoding="utf-8") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)  # 데이터 읽어온 자료 저장
    next(data)

    temp_high = []
    for row in data:
        if row[-1] != '':
            row[0] = row[0].replace('\t', '')  #년도앞 탭(공백) 제거
            if row[0].split('-')[1] == '08':
                temp_high.append(float(row[-1]))

print(len(temp_high))
# print(temp_high)
plt.rc('font', family='Malgun Gothic')
plt.title("서울의 8월 최고기온 분포")
plt.hist(temp_high, bins=100, color='r')
plt.xlabel('온도')
plt.ylabel('빈도수')
plt.show()
```
