```python
import pydataset as pyds
import pandas as pd

# 미국 환경 보호국의 mpg(mile per gallon) - 연비
mpg = pyds.data('mpg')
mpg

# csv 파일로 저장하기(객체 이름.to_csv()), index=False(unnamed가 삭제됨)
mpg.to_csv("./datas/mpg.csv", index=False)

# csv 파일 읽기(pd.read_csv())
mpg = pd.read_csv("./datas/mpg.csv")
mpg
```

```python
# head() - 처음 5행 출력
mpg.head()

# head() - 뒤 5행 출력
mpg.tail()

# info() - 요약 정보
mpg.info()

# describe() - 요약 통계
mpg.describe()
```

```python
# 통합 연비 - 칼럼명 추가
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
# mpg.head()

# print(type(mpg['total']))
# mpg.info()
print(mpg.describe())

# 통합 연비 평균 - 1. 총합 / 개수, 2. mean()
print(len(mpg))
# sum(mpg['total']) / len(mpg)
# mpg['total'].mean()
print(mpg['total'].describe())
# print(mpg['model'][:101])
```

```python
# 히스토그램 - 빈도, 분포를 알수 있는 그래프
# DataFrame 클래스에 plot 속성으로 그래프를 그릴 수 있음
# 분석 - 1.평균, 중앙값이 약 20이다. 2. 통합 연비가 20~22.5사이의 자동차 모델이 가장 많다.
#  3. 연비가 25를 넘가나 15보다 작은 자동차는 별로 없다.
mpg['total'].plot.hist()
```
