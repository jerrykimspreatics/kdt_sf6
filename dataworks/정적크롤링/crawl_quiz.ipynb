### 실습2. 전자신문 메인 기사 크롤링

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.etnews.com/20240820000076"
resp = requests.get(url)
# print(resp.text)
html = BeautifulSoup(resp.text, 'html.parser')
# print(html)

# title = html.find('div', class_='article_title')
title = html.select_one('div.article_title')
print(title.text.strip())

# time = html.find('time', class_="date")
time = html.select_one('time.date')
print(time.text.strip())
```

```python
# 본문 내용
# body_content = html.select('div.article_txt > p')
# print(body_content)

# for content in body_content:
#   print(content.text.strip())

body = html.find('div', class_='article_txt')
# print(body)

body_p = body.find_all('p')
# print(body_p)

for p in body_p:
  print(p.text)
```

### 실습3 환율정보 크롤링하기

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://finance.naver.com/marketindex/")
soup = BeautifulSoup(res.text, "html.parser")

# market1 검색
market1 = soup.select_one("div.market1")
# print(ul)

# 통화
exchange = market1.select_one('span.blind')
# print(exchange.text)

# 환율 지수
value = market1.select_one('span.value')
# print(value.text)

# 모든 환율 검색
all_li = market1.select("ul.data_lst > li")
# # print(all_li)
for li in all_li:
  exchange = li.select_one('span.blind')
  value = li.select_one('span.value')
  # text 대신 string도 가능함 - split(구분기호)은 리스트로 반환
  print(exchange.string.split(" ")[1], value.string)
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://finance.naver.com/marketindex/"
fund = requests.get(url)

market1 = BeautifulSoup(fund.text, 'html.parser')
# 통화
exchange = market1.select_one('span.blind')
print(exchange.text)
# 환율 지수
value = market1.select_one('span.value')
print(value.text)
# 모든 환율 검색
all_li = market1.select("ul.data_lst > li")
print(all_li)
for li in all_li:
  exchange = li.select_one('span.blind')
  value = li.select_one('span.value')
  # text 대신 string도 가능함 - split(구분기호)은 리스트로 반환
  print(exchange.string.split(" ")[0], value.string)
```
