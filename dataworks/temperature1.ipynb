```python
 # 최고 기온만 리스트에 담기
import csv
import matplotlib.pyplot as plt

# with open("datas/temp_seoul.csv", "r", encoding="utf-8") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)  # 데이터 읽어온 자료 저장
    next(data)

    temp_high = []
    for row in data:
        if row[-1] != '':
            row[0] = row[0].replace('\t', '')  #년도앞 탭(공백) 제거
            if int(row[0].split('-')[0]) >= 1980: #연도를 정수화 비교
                temp_high.append(float(row[-1]))

print(len(temp_high))
# print(temp_high)
plt.rc('font', family='Malgun Gothic')
plt.title("1980년 이후 서울의 최고기온")
plt.plot(temp_high, color='r')
plt.show()
```

```python
 # 최고 기온(최대값)과 해당 날짜 구하기
import csv

# with open("datas/temp_seoul.csv", "r", encoding="utf-8") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)  # 데이터 읽어온 자료 저장
    next(data)

    max_temp = -99  #최대값 설정
    max_date = ""   #최고기온 날짜, row[0]
    for row in data:
        if row[-1] != '':
            row[-1] = float(row[-1])
            row[0] = row[0].replace('\t', '')  #'\t'를 제거
            if row[-1] > max_temp:  #각각의 최고 기온과 최대값과 비교
                max_temp = row[-1]
                max_date = row[0]   #최고 기온 날짜

# print(f'최고 기온:{max_temp}, 날짜:{max_date.strip()}')
print(f'최고 기온:{max_temp}, 날짜:{max_date}')
```

 #### 내 생일의 최고 기온과 최저 기온 분석 및 시각화

```python
import csv
import matplotlib.pyplot as plt

# with open("datas/temp_seoul.csv", "r", encoding="utf-8") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)  # 데이터 읽어온 자료 저장
    next(data)

    temp_high = []  #최고 기온
    temp_low =[]    #최저 기온
    
    for row in data:
        if row[-1] != '':
            row[0] = row[0].replace('\t', '')  #년도앞 탭(공백) 제거
            if row[0].split('-')[0] >= '1980': #연도를 정수화 비교
                if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '27':
                    temp_high.append(float(row[-1]))  #최고 기온
                    temp_low.append(float(row[-2]))   #최저 기온

print(len(temp_high))
print(len(temp_low))
# print(temp_high)

plt.rc('font', family='Malgun Gothic')
plt.title("내 생일의 서울의 최고 기온과 최저 기온", pad=20)

plt.plot(temp_high, color='r', label='high')
plt.plot(temp_low, color='b', label='low')
plt.legend()
plt.xlabel('연도별 기온')
plt.ylabel('온도')
plt.show()
```
