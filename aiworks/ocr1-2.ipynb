```python
# 뉴스 기사 이미지 준비
import cv2
import matplotlib.pyplot as plt

# 이미지를 흑백으로 불러오기
img = cv2.imread("./source/news.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# tesseract 환경 설정
from pytesseract import pytesseract

# r"C:\Users\kiyon\AppData\Local\Programs\Tesseract-OCR\tesseract.exe"
pytesseract.tesseract_cmd = "C:/Users/kiyon/AppData/Local/Programs/Tesseract-OCR/tesseract.exe"

# 이미지를 텍스트로 추출
text = pytesseract.image_to_string(img, lang="kor+eng")
print(text)
```
