```python
import cv2
import matplotlib.pyplot as plt
from pytesseract import pytesseract

img = cv2.imread("source/sonata1.png", cv2.IMREAD_GRAYSCALE)

# 이진 이미지로 변경
_, bin_img = cv2.threshold(img, 35, 255, cv2.THRESH_BINARY)

plt.imshow(bin_img, cmap='gray')
plt.axis('off')
plt.show()

# tesseract 사용
pytesseract.tesseract_cmd = \
        "C:/Users/kiyon/AppData/Local/Programs/Tesseract-OCR/tesseract.exe"
        # r"C:\Users\kiyon\AppData\Local\Programs\Tesseract-OCR\tesseract.exe"
text = pytesseract.image_to_string(bin_img, lang="kor")
print(text)
```
