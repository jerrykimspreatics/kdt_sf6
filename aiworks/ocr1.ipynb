```python
import cv2
from pytesseract import pytesseract
import matplotlib.pyplot as plt

# 이미지를 흑백으로 불러오기
img = cv2.imread("source/news1.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# 테세렉트 환경 설정
pytesseract.tesseract_cmd = \
        r"C:\Users\kiyon\AppData\Local\Programs\Tesseract-OCR\tesseract.exe"

# 이미지를 텍스트로 읽기
text = pytesseract.image_to_string(img, lang="kor+eng")
print(text)
# print(text.replace(" ", ""))
```
